```python
print(f"Current Waste Value: R{int(total_waste_value):,}")
print(f"Target Waste Value: R{int(target_waste_value):,}")
recovery_opportunity = total_waste_value - target_waste_value
print(f"Recovery opportunity: R{int(recovery_opportunity):,}")
```
